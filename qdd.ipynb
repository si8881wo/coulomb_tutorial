# Setup

We think about the typicall quantum dot setup, where we have a dot connected to two leads.

(here could be an image of the setup :) )

In this system we have many parameters to play around with.
We will here work with

- bias voltage $V_b$
- gate voltage $V_g$
- lead temperatures $T_{L,R}$
- charging energy $U$.

Depending on these we want to investigate the following quantities:

- current $I$
- (differential) conductance $\frac{dI}{dV_b}$.

Let's talk about them step by step.

## Bias and gate voltage

These leads are characterized by the so called *chemical potential*.
It basically tells us how many electrons are in the leads.
The difference in chemical potential is called *bias*.
It is an elecrical voltage between the two leads and can therefore be easily controled experimentally.
This is one of the parameters we will sweep.

The *gate voltage* can be used to tune the energy level of the dot.
That means by controling this we effectively move the dot level up and down.
Again this is a voltage and we can easlily control it experimentally.
This is the other parameter we will sweep.

## Lead temperatures

We can think of the leads as big reservoirs of electrons, that have a certain energy distribution, called *Fermi-Dirac-distribution*.
The electrons will not all have the same energy, but some will have more energy than others.
If the temperature is absolute zero, then the electrons will be evently distributed up to the *chemical potential* that we talked about earlier.
There will be no electrons at higher energy.
If we increase the temperature, some of the electrons will have a little more energy.
Temperature "smears out" the sharp drop in electron energy.
We can have the same temperature in both leads or different temperatures.
This can lead to interesting effects.

## Coulomb interaction

Quantum dots are very small structures, so small that we can only fit one electron per spin on it.
Because they are so small, an electron can "feel" if an electron with a different spin is already on the dot, when it wants to hop to the dot.
It takes extra energy to put the other spin electron on the dot if it is already occupied.
That energy is called *charging energy* and comes from the *Coulomb interaction* between the electrons.
Remember, electrons are negatively charged particles, so they repel each other.

## Current and conductance

From the parameters we talked about above we want to calculate and investigate the *current* and *(differential) conductanc*

If we put a *bias voltage* across the dot (i.e. the chemical potential is big on one side and low on the other), the electrons will try to get through the dot.
The properties of the dot then determine how many electrons can be transported.
The number of electrons that hop to the other side per time is the *current*.

The *(differential) conductance* tells us how much the *current* changes, if we change the *bias voltage* a little bit.
It is the derivative of the *current* after the *bias voltage* and can sometimes give us more insights into what is going on or make certain things more visible.
